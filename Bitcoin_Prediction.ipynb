```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv(r"/content/BTC-USD_Historical_Data.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.drop_duplicates()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
date=dataset['Date'].str.split('-',expand=True)
dataset['Day']=date[0].astype('int')
dataset['Month']=date[1].astype('int')
dataset['Year']=date[2].astype('int')
dataset.drop("Date",inplace=True,axis=1)
dataset
```

```python
dataset['Day']=dataset.pop('Day')
dataset['Month']=dataset.pop('Month')
dataset['Year']=dataset.pop("Year")
```

```python
dataset
```

```python
dataset = dataset[['Year', 'Month', 'Day'] + [col for col in dataset.columns if col not in ['Year', 'Month', 'Day']]]
dataset
```

```python
dataset['Close']=dataset.pop('Close')
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
features=dataset.drop("Close",axis=1)
features_scaled=ss.fit_transform(features)
dataset.loc[:, features.columns] = features_scaled
#dataset.to_csv('Amazon_Stock_Prediction.csv', index=True)
dataset
```

```python
from sklearn.model_selection import train_test_split
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
import xgboost as xgb
model=xgb.XGBRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# Calculate evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R² Score: {r2}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
plt.plot(dataset.index[-len(y_test):], y_test, label="Actual Price", color='blue')
plt.plot(dataset.index[-len(y_test):], y_pred, label="Predicted Price", color='red')
plt.xlabel("Date")
plt.ylabel("Bitcoin Price")
plt.title("Bitcoin Price Prediction using XGBoost")
plt.legend()
plt.show()
```

```python
model1.compile(optimizer='adam',loss="mean_absolute_error")
metrics=[mean_absolute_error]
model1.fit(x_train,y_train,epochs=200,batch_size=32,validation_data=(x_test,y_test))
```

```python
y_pred_ann=model1.predict(x_test)
```

```python
def create_sequences(dataset, time_steps=50):
    x_seq, y_seq = [], []
    for i in range(len(dataset) - time_steps):
        x_seq.append(dataset[i:i + time_steps])
        y_seq.append(dataset[i + time_steps,0])
    return np.array(x_seq), np.array(y_seq).reshape(-1,1)

time_steps = 50  # Number of past days used to predict the next day
x_seq, y_seq = create_sequences(features_scaled, time_steps)

# Split data into train (80%) and test (20%)
split = int(0.8 * len(x_seq))
x_seq_train, y_seq_train = x_seq[:split], y_seq[:split]
x_seq_test, y_seq_test = x_seq[split:], y_seq[split:]
```

```python
from tensorflow.keras.layers import GRU
from tensorflow.keras.callbacks import EarlyStopping
model3=Sequential()
model3.add(GRU(128,return_sequences=True,input_shape=(time_steps,7)))
model3.add(GRU(64,return_sequences=False,activation='relu'))
model3.add(Dense(25,activation='relu'))
model3.add(Dense(1))
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
print(model3.summary())
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)


model3.compile(optimizer='adam',loss='mean_squared_error')
model3.fit(x_seq_train,y_seq_train,epochs=100,batch_size=32,validation_data=(x_seq_test,y_seq_test))
```
